### Introduction
The purpose of this project is to conduct an exploratory data analysis on the 2021-2022 Champions League season ⚽️. The primary objective is to present intriguing statistical information through effective data visualization techniques.

## UCL, the most amazing time for football fans
The Union of European Football Associations (UEFA) organizes the annual UEFA Champions League (UCL), which features Europe's top-division football clubs competing for the championship title. The tournament follows a round-robin group stage, which leads to a double-legged knockout format and a single-legged final. As one of the most esteemed football tournaments worldwide, the Champions League draws the national league champions (and, in some countries, one or more runners-up) from their respective associations. For passionate football fans, the arrival of the Champions League sparks an unparalleled surge of emotions and enthusiasm for their team, making it a highly anticipated occasion.

### Goal contributions per minutes played

```python
min_played_threshold = 180
df_Gc = df_key_stats[df_key_stats['minutes_played'] >= min_played_threshold].copy()
df_Gc['gc_per_min'] = (df_Gc['goals'] + df_Gc['assists']) / df_Gc['minutes_played']
df_top10Gc = df_Gc.groupby(['player_name', 'club', 'position'])['gc_per_min'].mean().reset_index().sort_values('gc_per_min', ascending=False)[:10]
sns.set_style('whitegrid')
sns.barplot(x='player_name', y='gc_per_min', data=df_top10Gc, palette='rocket')
plt.title('Top 10 Players by Goal Contributions per Minute Played', fontsize=20)
plt.xlabel('Player Name', fontsize=15)
plt.ylabel('Goal Contributions per Minute Played', fontsize=15)
plt.xticks(rotation=45, fontsize=12)
plt.show()
```
